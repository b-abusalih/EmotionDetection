```python
df = pd.read_csv("textprobe.csv")
fig= sns.displot(data=df,palette="Set2")
fig.set_axis_labels( 'textprobe','Count')
sns.move_legend(fig, "lower center",
    bbox_to_anchor=(.5, 0.95), ncol=4, title=None, frameon=False,)
fig.savefig('textprobe.jpg', format='jpeg', dpi=600)
```
